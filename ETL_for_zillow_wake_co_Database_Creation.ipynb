```python
# Import Dependencies
import pandas as pd
```

```python
# Load in properties_2016
zillow_wake_co = "Resources/zillow_wake_co.csv"
```

```python
# Read zillow wake co into dataframe
zillow_wake_co = pd.read_csv(zillow_wake_co)
zillow_wake_co_df = zillow_wake_co
#zillow_wake_co_df.head()
```

```python
# Check to see df loaded correctly
zillow_wake_co_df.head()
```

```python
# Use describe 
zillow_wake_co_df.describe
```

```python
# Check to see the data types of the df
zillow_wake_co_df.dtypes
```

```python
# Check for missing values
zillow_wake_co_df.count()
```

```python
# Drop 'props.contingentListingType' & 'props.hasImage' 
zillow_wake_co_df_clean = zillow_wake_co_df.drop(['props.contingentListingType', 'props.hasImage'], axis=1)
zillow_wake_co_df_clean.head()
```

```python
# Drop unnamed values from df
zillow_wake_co_df_clean.drop(zillow_wake_co_df_clean.columns[zillow_wake_co_df_clean.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
# Check for missing values
zillow_wake_co_df_clean.isnull().sum()
```

```python
#Drop the nulls
zillow_wake_co_df_clean=zillow_wake_co_df_clean.dropna()
```

```python
# Check for missing values
zillow_wake_co_df_clean.isnull().sum()
```

```python
# Check the data
zillow_wake_co_df_clean.head()
```

```python
# Split the address into additional home address, city, state, and zip columns
zillow_wake_co_df_clean[['props.mail_address', 'props.city', 'props.state']]= zillow_wake_co_df_clean['props.address'].str.split(',', expand=True)
```

```python
# Strip the white space from the props.state column
zillow_wake_co_df_clean['props.state']=zillow_wake_co_df_clean['props.state'].str.strip()
```

```python
# Separate the state and zip code into two columns 
zillow_wake_co_df_clean[['props.state', 'props.zipcode']]= zillow_wake_co_df_clean['props.state'].str.split(' ', expand=True)
```

```python
# Drop the original address column and reorder
zillow_wake_co_df_clean = zillow_wake_co_df_clean.reindex(columns=['props.dateSold', 'props.propertyType', 'props.lotAreaValue', 'props.mail_address', 'props.city','props.state', 'props.zipcode', 'props.variableData.text', 'props.variableData.type', 'props.zestimate', 'props.imgSrc', 'props.price', 'props.bedrooms', 'props.daysOnZillow', 'props.bathrooms', 'props.livingArea', 'props.country', 'props.currency', 'props.lotAreaUnit'])
```

```python
# Check the df
zillow_wake_co_df_clean.head()
```

```python
# Show the coorelation between living area to price
zillow_wake_co_df_clean.plot.scatter(x='props.livingArea', y='props.price')
```
