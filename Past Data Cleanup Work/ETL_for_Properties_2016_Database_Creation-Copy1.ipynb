```python
# Read properties_2016 into dataframe
properties_2016 = pd.read_csv((properties_2016), low_memory=False)
#properties_2016 = pd.read_csv(properties_2016)
properties_2016_df = properties_2016
properties_2016_df.head(5)
```

```python
# Cleaning issue with Object datatype columns
object_columns = properties_2016_df.select_dtypes(include=['object']).columns

#Display columns
print('\nobject columns:\n', object_columns)
```

```python
# Counting Nan/null in above columns
print("\nCount total Nan at each column in a Dataframe : \n\n", properties_2016_df.isnull().sum())
```

```python
# Drop columsn where Nan is dominant (approx. half or more of the data in row)
properties_2016_df = properties_2016_df.drop(['airconditioningtypeid', 'architecturalstyletypeid', 'basementsqft', 'buildingclasstypeid', 'decktypeid', 'finishedfloor1squarefeet', 'finishedsquarefeet12', 'finishedsquarefeet13', 'finishedsquarefeet15', 'finishedsquarefeet50', 'finishedsquarefeet6', 'fireplacecnt', 'garagecarcnt', 'garagetotalsqft', 'hashottuborspa', 'heatingorsystemtypeid', 'poolcnt', 'poolsizesum', 'pooltypeid10', 'pooltypeid2', 'pooltypeid7', 'propertyzoningdesc', 'regionidneighborhood', 'storytypeid', 'threequarterbathnbr', 'typeconstructiontypeid', 'unitcnt', 'yardbuildingsqft17', 'yardbuildingsqft26', 'numberofstories', 'fireplaceflag', 'taxdelinquencyflag', 'taxdelinquencyyear'], axis=1)
properties_2016_df
```

```python
# Drop rows where longitude is null/NaN
properties_2016_df = properties_2016_df.dropna(subset=['regionidzip'])
properties_2016_df
```

```python
#Split dataframe due to large size and gitlab limits
properties_2016_df1 = properties_2016_df.iloc[:535000]
properties_2016_df2 = properties_2016_df.iloc[535000:1070000]
properties_2016_df3 = properties_2016_df.iloc[1070000:1605000]
properties_2016_df4 = properties_2016_df.iloc[1605000:2140000]
properties_2016_df5 = properties_2016_df.iloc[2140000:2675000]
properties_2016_df6 = properties_2016_df.iloc[2675000:]
```
