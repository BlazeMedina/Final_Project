```python
## Import dependencies
import pandas as pd
import requests
import json
import time
from config import api_key
```

```python
# list of zip codes in Greater Triangle Area
zip_1 = ['27701','27705','27707','27713','27709','27703','27704','27587','27614','27571',
         '27597','27591','27545','27616','27604','27615','27613','27617','27612','27609']
zip_2 = ['27608','27605','27601','27607','27519','27523','27513','27511','27606','27518',
         '27539','27502','27562','27540','27526','27592','27529','27603','27231','27243']
zip_3 = ['27279','27516','27514','27510','27517','27527','27576','27542','27569','27520',
         '27577','27524','27504']

#set current list of zip codes wanted to retrieve to zc
zc= zip_2
```

```python
## Realtor.com API response for zip 
url = "https://realtor-search.p.rapidapi.com/property/search-sale"


headers = {
	"X-RapidAPI-Key": api_key,
	"X-RapidAPI-Host": "realtor-search.p.rapidapi.com"
}

# for loop to make requests from API from all zipcodes zc
for i in range(0, len(zc)):
    try:        
        querystring = {"location":zc[i],"search_within_x_miles":"0"}    
        response = requests.get(url, headers=headers, params=querystring)
        json_data = response.json()
        print(f"Successfully retrieved response for zip code: {zc[i]}")
    except:
        print(f"There was an error retrieving the json for zip code: {zc[i]}")
    
    #Save json file with zipcode in file name
    with open(f'realtor_{zc[i]}.json', 'w') as file_writer:
        json.dump(json_data, file_writer)

    #Wait at least 1 minute between calls 
    print("Waiting 10 seconds before next call request")
    for j in range(10):
        print("*", end=" ")
        time.sleep(1)
    print("")
    
#Let the user know that the requests have ended
print("API call requests have ended")
    
```
