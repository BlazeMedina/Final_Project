```python
import pandas as pd
from pathlib import Path
import json

# absolute path to json file
jsonpath = Path('realtor_27006.json')
jsonpath1 = Path('realtor_27012.json')
jsonpath2 = Path('realtor_27017.json')
jsonpath3 = Path('realtor_27018.json')
jsonarray = [jsonpath,jsonpath1,jsonpath2,jsonpath3]
# reading the json file
i=0
dframe=[]
for jsonpath in jsonarray:
     i=i+1
     with jsonpath.open('r', encoding='utf-8') as dat_f:
        dat = json.loads(dat_f.read())
        df =pd.json_normalize(dat['data'])
        dframe.append(df)
dframe       

result = pd.concat([dframe[i] for i in range(len(dframe))], axis=0)

result.to_csv('datafile30.csv', encoding='utf-8', index=False)

del result['branding']
result
```
