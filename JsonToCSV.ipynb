```python
import pandas as pd
from pathlib import Path
import json

# absolute path to json file
jsonpath = Path('realtor_27006.json')
jsonpath1 = Path('realtor_27012.json')
jsonpath2 = Path('realtor_27017.json')
jsonpath3 = Path('realtor_27018.json')

# reading the json file
with jsonpath.open('r', encoding='utf-8') as dat_f:
    dat = json.loads(dat_f.read())

# creating the dataframe
df = pd.json_normalize(dat['data'])


with jsonpath1.open('r', encoding='utf-8') as dat_f1:
    dat1 = json.loads(dat_f1.read())

# creating the dataframe
df1 = pd.json_normalize(dat1['data'])



with jsonpath2.open('r', encoding='utf-8') as dat_f2:
    dat2 = json.loads(dat_f2.read())

# creating the dataframe
df2 = pd.json_normalize(dat2['data'])



with jsonpath3.open('r', encoding='utf-8') as dat_f3:
    dat3 = json.loads(dat_f3.read())

# creating the dataframe
df3 = pd.json_normalize(dat3['data'])

pd.concat([
    pd.concat([df, df1]),
    pd.concat([df2, df3])]).to_csv('datafile4.csv', encoding='utf-8', index=False)
# converted a file to csv
df
```
